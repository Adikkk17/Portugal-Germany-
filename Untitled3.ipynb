```python
df = sb.events(match_id = 3788764)
df
```

```python
recball = df[df['type'] == 'Ball Recovery']
recball 
```

```python
recball = recball [['team', 'type' , 'minute', 'location']]
recball 
```

```python
import pandas as pd 
recball[['X', 'Y']] = recball.location.apply(pd.Series)
recball = recball.drop('location' , axis= 1)
recball 
```

```python
!pip install matplotlib seaborn pandas
```

```python
# Calculate the percentage of ball recoveries for each team
total_recoveries = len(recball)
team_counts = recball['Team'].value_counts()
team_percentages = (team_counts / total_recoveries) * 100
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as patches

# Create a football pitch
def create_pitch():
    fig, ax = plt.subplots(figsize=(12, 8))
    # Pitch Outline & Centre Line
    plt.plot([0, 0, 120, 120, 0], [0, 80, 80, 0, 0], color="black")
    plt.plot([60, 60], [0, 80], color="black")
    
    # Left Penalty Area
    plt.plot([18, 18], [62, 18], color="black")
    plt.plot([0, 18], [62, 62], color="black")
    plt.plot([18, 0], [18, 18], color="black")
    
    # Right Penalty Area
    plt.plot([120, 102], [62, 62], color="black")
    plt.plot([102, 102], [62, 18], color="black")
    plt.plot([102, 120], [18, 18], color="black")
    
    # Left 6-yard Box
    plt.plot([0, 6], [50, 50], color="black")
    plt.plot([6, 6], [50, 30], color="black")
    plt.plot([6, 0], [30, 30], color="black")
    
    # Right 6-yard Box
    plt.plot([120, 114], [50, 50], color="black")
    plt.plot([114, 114], [50, 30], color="black")
    plt.plot([114, 120], [30, 30], color="black")
    
    # Prepare Circles; 10 yard circle at centre and 1 yard circle at penalty spots
    centreCircle = plt.Circle((60, 40), 10, color="black", fill=False)
    centreSpot = plt.Circle((60, 40), 1, color="black")
    leftPenSpot = plt.Circle((12, 40), 1, color="black")
    rightPenSpot = plt.Circle((108, 40), 1, color="black")
    
    # Draw Circles
    ax.add_patch(centreCircle)
    ax.add_patch(centreSpot)
    ax.add_patch(leftPenSpot)
    ax.add_patch(rightPenSpot)
    
    # Draw Arcs
    leftArc = patches.Arc((12, 40), height=16, width=16, angle=0, theta1=308, theta2=52, color="black")
    rightArc = patches.Arc((108, 40), height=16, width=16, angle=0, theta1=128, theta2=232, color="black")
    
    ax.add_patch(leftArc)
    ax.add_patch(rightArc)
    
    return fig, ax

# Plot the data on the pitch
fig, ax = create_pitch()

# Use seaborn to plot the points
sns.scatterplot(data=recball, x='X', y='Y', hue='Team', style='Team', s=100, ax=ax, palette='coolwarm', alpha=0.6)

# Adding title and labels
plt.title('Ball Recoveries with Percentages')
plt.xlabel('X Coordinate')
plt.ylabel('Y Coordinate')

# Add text for percentages
for team, percentage in team_percentages.items():
    plt.text(125, 80 - team_counts.index.get_loc(team) * 5, f'{team}: {percentage:.2f}%', 
             fontsize=12, color='black')

# Show the plot
plt.show()

```
